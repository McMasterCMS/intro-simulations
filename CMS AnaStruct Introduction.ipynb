## Welcome to AnaStruct! Code and documentation here:
https://anastruct.readthedocs.io/en/latest/

AnaStruct is an open source python library that allows for beam element simulations. Setting up a simulation involves the following:
1. Adding system beam elements.
2. Defining fixed point constraints.
3. Defining applied forces.
4. Running the simulation to solve for displacement, reaction forces and bending moments. 

While there already exists software that performs a much more refined and sophisticated structural analysis such as Autodesk Inventor or ANSYS, learning to use this library poses a few key benefits...

1. Performing basic structural analysis...duh
2. Learning to problem solve in using someone else's code by reading their documentation.
3. Applying optimization techniques scriptable in python (such as machine learning using Scikit-learn, PyTorch, Tensorflow-keras etc...)
